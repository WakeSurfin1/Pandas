```python
import pandas as pd

df = pd.read_csv("C:/Users/jtcUser/Documents/Systems/Training/PythonCourse/Section4/files/AB_NYC_2019.csv")

## df.head()
```

```python
## slicing columns
df[["host_name", "neighbourhood_group"]]
```

```python
## Filtering on rows (mask filtering)
## return all columns of the dataframe
df[df.host_name == "Taz"]
```

```python
## returns the index only
df.host_name == "Taz"
```

```python
## sum the number of rows the fitlered dataset returns
(df.host_name == "Taz").sum()
```

```python
## masking
## create a boolean mask
mask = df.host_name == "Taz"
## index the dataframe using the boolean mask from above
df[mask].head(3)
```

```python
## create a boolean mask based on 2 conditions (and) 
quick_and_cheap = (df.price < 100) & (df.minimum_nights < 3)
## output the sum of rows that satify the two conditions
quick_and_cheap.sum()
```

```python
## filter the dataframe retaining only the records where mask is true
df[quick_and_cheap].head(3)
```

```python
## create a boolean mask based on 2 conditions (or)
reviews_consistent = df[(df.reviews_per_month > 3) | (df.number_of_reviews > 50)]
reviews_consistent.head(3)
```

```python
## the loc function is used for label-based indexing and slicing of a DataFrame. 
## It allows you to access rows and columns based on their labels.

## .loc - filtering columns and rows together
## mask = df.host_name == "Taz"

df.loc[mask, ["name", "host_name"]]
```

```python
## return all columns
df.loc[mask, :]
```

```python
## The iloc function s used for integer-based indexing and slicing of a DataFrame. 
## It allows you to access rows and columns based on their numerical positions rather than their labels.

## .iloc filtering based on index
df2 = df.set_index("id")
df2.iloc[0, :]
```

```python
## iloc with slicing
df2.iloc[1:4, 6:]
```

```python
## Note: between function - include data where element values are between the specified parameters
## Mask helpers
df.loc[df.price.between(100, 200), "price"].head()
```

```python
## simple filter
## check each element of the dataframe to see if it's equal to "John"
df == "John"
```

```python
## The any function is used to determine whether any elements in a DataFrame or Series evaluate to True. 
## It returns a single boolean value that indicates whether at least one element in the given data structure is True.

## .any()
## search for columns that contain the string "John"
(df == "John").any()
```

```python
## .any(axis=1)
## search for the string "John" any where in a row
## return the rows that have the string "John" any cell in the row
(df == "John").any(axis=1)
```

```python
## Views vs Copy
df2 = df.copy()
```
