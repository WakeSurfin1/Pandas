```python
## Indexing - the unique value that identified each row. Default is counting from 0 up.  
## the index can be set to a column in the data set.

df2 = df.set_index("id")
df2.head()
```

```python
## data can be referenced by index value
df2.name[2539]
```

```python
## How to change the index using groupby() - The index becomes "room_type"
df3 = df.groupby("room_type").mean()
df3.reset_index()
```

```python
## reset index back to pandas default
df3.reset_index(drop=True)
```

```python
## Sorting by the index
df3.sort_index(ascending=False)
```

```python
## Sort by column name
df.sort_values("host_name").head(3)
```

```python
## sort by multiple columns
df.sort_values(["neighbourhood_group", "host_name"])
df.head(3)
```

```python
## sort by multiple columns using ascending
df.sort_values(["neighbourhood_group", "host_name"], ascending=[False, True]).head(3)
```

```python
dfp = df.sort_values("price", ascending=False)
dfp[["id", "host_name", "price"]].head(5)
```

```python
## Rank is similar to sort but rank also adds collision detection
dfp["price_rank"] = dfp.price.rank(method="average", ascending=False)
dfp[["id", "host_name", "price", "price_rank"]].head(5)
```
